```python
columns_to_keep = [
    'Date', 'Time', 'Comp', 'Round', 'Venue', 'Result', 'GF', 'GA', 
    'Opponent', 'Formation', 'Referee', 'Sh', 'SoT', 'Dist', 'FK', 
    'PK', 'PKatt', 'Season', 'Team'
]

columns_to_drop = [
    'Day', 'xG', 'xGA', 'Poss', 'Attendance', 'Captain', 'Opp Formation',
    'Match Report', 'Notes', 'Date.1', 'Time.1', 'Comp.1', 'Round.1', 
    'Day.1', 'Venue.1', 'Result.1', 'GF.1', 'GA.1', 'Opponent.1', 
    'Gls', 'SoT%', 'G/Sh', 'G/SoT', 'xG.1', 'npxG', 'npxG/Sh', 
    'G-xG', 'np:G-xG', 'Match Report.1', 'ï»¿Date'
]
```

```python
matches = matches[columns_to_keep]

matches_columns = matches.columns
print(matches_columns)
```

```python
matches.head()
```

```python
matches = matches[matches['Comp'] == 'Premier League']
matches.head()
```

```python
matches.columns = [c.lower() for c in matches.columns]

matches.head()
```

```python
matches['round'].value_counts()
```

```python
matches.dtypes
```

```python
matches['date'] = pd.to_datetime(matches['date'], format='mixed', dayfirst=True)
matches['gf'] = matches['gf'].astype(float)
matches['ga'] = matches['ga'].astype(float)
matches['season'] = matches['season'].astype(str)
matches['season'] = matches['season'].str.split('-').str[0].astype(int)
```

```python
matches.dtypes
```

```python
# Convert the venue into a venue code

matches['venue_code'] = matches['venue'].astype('category').cat.codes

# Do the same for the opponents

matches['opp_code'] = matches['opponent'].astype('category').cat.codes
```

```python
# Convert time into int (maybe teams perform better at certain times of day)
matches['hour'] = matches['time'].str.replace(':.+', '', regex=True).astype('int')
```

```python
# Do the same for date too
matches['day_code'] = matches['date'].dt.day_of_week
```

```python
# Create a target. Whether the team won or not
matches['target'] = (matches['result'] == 'W').astype('int')
```

```python
matches
```

```python
# Import random forest model from scikit learn
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = matches[matches['date'] < '2024-01-01']
```

```python
test = matches[matches['date'] > '2024-01-1']
```

```python
predictors = ['venue_code', 'opp_code', 'hour', 'day_code']
```

```python
rf.fit(train[predictors], train['target'])
```

```python
preds = rf.predict(test[predictors])
```

```python
# How we determine the accuracy of the model
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test['target'], preds)
acc
```

```python
combined = pd.DataFrame(dict(actual=test['target'], prediction=preds))
```

```python
pd.crosstab(index=combined['actual'], columns=combined['prediction'])
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test['target'], preds)
```

```python
# Create more predictors to improve accuracy
grouped_matches = matches.groupby('team')
```

```python
group = grouped_matches.get_group('Manchester_City')
group
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values('date')
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ['gf', 'ga', 'sh', 'sot', 'dist', 'fk', 'pk', 'pkatt']
new_cols = [f'{c}_rolling' for c in cols]
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby('team').apply(lambda x: rolling_averages(x, cols, new_cols), include_groups=False)
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data['date'] < '2024-01-01']
    test = data[data['date'] > '2024-01-1']
    rf.fit(train[predictors], train['target'])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test['target'], prediction=preds), index=test.index)
    precision = precision_score(test['target'], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
precision
```
